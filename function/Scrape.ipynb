```python
url ="https://sjp.moj.gov.sa/Filter?isFilterButtonClicked=True"
```

```python
r = requests.get(url, headers=headers,verify=False)
print(r.status_code)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.title.text)
```
